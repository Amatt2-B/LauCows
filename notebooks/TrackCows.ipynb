# Track Cows
This script loads images from a directory and generates the prediction for each 
one of them.

```python
from ultralytics import YOLO
from dotenv import load_dotenv
from pathlib import Path
from PIL import Image
import torch
import cv2
import os
```

**IMPORTANT!!!** 

This notebook loads the paths to the dataset using a .env file
if you don't have it already, create it on the notebooks/ directory and fill the 
variables from .template.env

```python
load_dotenv()
# Change this directories to the location where you have the images to track
PWD = Path(os.getenv('COW_DETECT_PWD'))
ALL_IMG_DIR = Path(os.getenv('COW_DETECT_ALL_COWS_DIR')).joinpath('images')
testimgs = os.listdir(PWD.joinpath('test/images/'))
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# Load model
model = YOLO(Path('../models/best.pt'))
model.to(device);
```

```python
# Take all the image names to make predictions
allImgs = [img for img in os.listdir(ALL_IMG_DIR)]
```

```python
# Iterate all images
for imageName in allImgs:
    img = cv2.imread(ALL_IMG_DIR.joinpath(imageName))

    # Generate predictions
    preds = model.predict(img)

    # Post process the image to rgb format
    plot = preds[0].plot()
    plot = cv2.cvtColor(plot, cv2.COLOR_BGR2RGB)
    
    # Save the image on the tmp folder
    Image.fromarray(plot).save(Path('../tmp', imageName))
    
```
